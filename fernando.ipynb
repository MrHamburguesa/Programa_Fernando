```python
import pandas as pd
```

```python
df = pd.read_excel('SEROLOGIA MAYO 2022.xlsx')
df['Descripción'] = df['Descripción'].fillna('No hubo notificación')
```

```python
datos_totales = []
etareo = sorted(df.Etáreo.unique())
descripciones = sorted(df.Descripción.unique())

for tipo_de_dato in descripciones:
    entradas = []
    numeros_totales = df[df['Descripción'] == tipo_de_dato].shape[0]
    entradas.append(numeros_totales)
    for eta in etareo:
        df_util = df[(df['Descripción'] == tipo_de_dato) & (df['Etáreo'] == eta)]
        numero = df_util.shape[0]
        entradas.append(numero)
    
    datos_totales.append(entradas)

df_nuevisima = pd.DataFrame(datos_totales, columns = ['Total'] + etareo, index = descripciones)
```
